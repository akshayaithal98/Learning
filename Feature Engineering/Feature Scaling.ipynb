```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
#perform scaling only to numerical variables
```

## Standardization

```python
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv("DS2_C5_S4_Wine_Data_Practice-Copy1.data",header=None).drop(0,axis=1)
```

```python
df.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(df.iloc[:,:-1],df[13],test_size=0.2,random_state=1)
```

```python
x_train.describe()
```

```python
scaler=StandardScaler()
```

```python
x_train_s=pd.DataFrame(scaler.fit_transform(x_train),columns=x_train.columns)
x_test_s=pd.DataFrame(scaler.transform(x_test),columns=x_test.columns)
```

```python
x_train_s.head()
```

```python
#scaled down to 0 mean and unit std for all variables
#min and max values across variables vary
x_train_s.describe()
```

```python
x_train[1].hist()
```

```python
#we can see after scaling original shape of the distribution is preserved
x_train_s[1].hist()
```

```python
#for demo purpose visualizing only 3 variables how they were before and after scaling
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# before scaling
ax1.set_title('Before Scaling')
sns.kdeplot(x_train[1], ax=ax1)
sns.kdeplot(x_train[2], ax=ax1)
sns.kdeplot(x_train[3], ax=ax1)

# after scaling
ax2.set_title('After Standard Scaling')
sns.kdeplot(x_train_s[1], ax=ax2)
sns.kdeplot(x_train_s[2], ax=ax2)
sns.kdeplot(x_train_s[3], ax=ax2)
plt.show()
```

## Normalization

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
x_train,x_test,y_train,y_test=train_test_split(df.iloc[:,:-1],df[13],test_size=0.2,random_state=1)
```

```python
scaler=MinMaxScaler()
```

```python
x_train_s=pd.DataFrame(scaler.fit_transform(x_train),columns=x_train.columns)
x_test_s=pd.DataFrame(scaler.transform(x_test),columns=x_test.columns)
```

```python
x_train.describe()
```

```python
#scaled down between 0 to 1
#mean is not centred and variance varies acros variables
x_train_s.describe()
```

```python
plt.figure(figsize=(12,5))
plt.subplot(121)
sns.histplot(data=x_train,x=5)
plt.subplot(122)
sns.histplot(data=x_train_s,x=5)
```

```python
#so the shape of the distribution is not varied much
#so the shape may or may not vary
```

```python
#for demo purpose visualizing only 3 variables
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# before scaling
ax1.set_title('Before Scaling')
sns.kdeplot(x_train[1], ax=ax1)
sns.kdeplot(x_train[2], ax=ax1)
sns.kdeplot(x_train[3], ax=ax1)

# after scaling
ax2.set_title('After Min-Max Scaling')
sns.kdeplot(x_train_s[1], ax=ax2)
sns.kdeplot(x_train_s[2], ax=ax2)
sns.kdeplot(x_train_s[3], ax=ax2)
plt.show()
```

## Robust Scaling

```python
#this is highly robust to outliers
#so use this when therea re outliers
```

```python
from sklearn.preprocessing import RobustScaler
```

```python
x_train,x_test,y_train,y_test=train_test_split(df.iloc[:,:-1],df[13],test_size=0.2,random_state=1)
scaler=RobustScaler()
x_train_s=pd.DataFrame(scaler.fit_transform(x_train),columns=x_train.columns)
x_test_s=pd.DataFrame(scaler.transform(x_test),columns=x_test.columns)
```

```python
x_train.describe()
```

```python
#centres the median at 0
#variance ,min,max values vary across variables
x_train_s.describe()
```

```python
#distribution may vary but here its same
sns.histplot(data=x_train,x=2,bins=30)
plt.show()
sns.histplot(data=x_train_s,x=2,bins=30)
plt.show()
```
