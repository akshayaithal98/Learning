```python
df['mileage']=(df["City_Mileage_km_litre"]+df["Highway_Mileage_km_litre"])/2
```

```python
df["mileage"]
```

```python
corr1=df["mileage"].corr(df["Displacement"])
corr1
```

```python
corr2=df["mileage"].corr(df["Fuel_Tank_Capacity_litre"])
corr2
```

```python
corr3=df["mileage"].corr(df["Minimum_Turning_Radius"])
corr3
```

```python
plt.figure(figsize=(15,5))


plt.subplot(131)
plt.xlabel("mileage")
plt.ylabel("Displacement")
plt.scatter(df["mileage"],df["Displacement"],label="Corr="+str(corr1),color="red")
plt.legend()

plt.subplot(132)
plt.xlabel("mileage")
plt.ylabel("Fuel_Tank_Capacity_litre")
plt.scatter(df["mileage"],df["Fuel_Tank_Capacity_litre"],label="Corr="+str(corr2),color="blue")
plt.legend()
plt.subplot(133)
plt.xlabel("mileage")
plt.ylabel("Minimum_Turning_Radius")
plt.scatter(df["mileage"],df["Minimum_Turning_Radius"],label="Corr="+str(corr3),color="yellow")
plt.legend()
plt.show()
```

```python
#mileage is negatively correlated with displacement,fuel tank capacity and turning radius with correlation of
#-0.6894744971164131,-0.7145034450182555,-0.35620940763419046 repectively while fuel tank capacity is highly negative correlated.
#this means that increase in fuel tank capacity have the inverse effect on mileage as decreases in the mileage as with other factors.
```
