```python
import numpy as np
import pandas as pd
import scipy.stats as stat
import matplotlib.pyplot as plt
import seaborn as sb
import math
import warnings as w
w.filterwarnings("ignore")
```

# Task1

```python
mean=1332
sd=725
```

```python
#a
x=2000
zscore=(x-mean)/sd
p=1-stat.norm.cdf(zscore)
print(p,"or",p*100,"%")
```

```python
#b
x=0
zscore=(x-mean)/sd
p=stat.norm.cdf(zscore)
print(p,"or",p*100,"%")
```

```python
#c
x1=100
x2=700
z1=(x1-mean)/sd
z2=(x2-mean)/sd
stat.norm.cdf(z2)-stat.norm.cdf(z1)
```

```python
lower=mean-3*sd
upper=mean+3*sd
data=np.arange(lower,upper)
plt.xlabel("tax returns")
plt.ylabel("probability")
plt.plot(data,stat.norm.pdf(data,mean,sd))
plt.show()
```

```python
#this graph clearly shows that within +-1SD from mean probability of having our data is high
```

# Task2

```python
df=pd.read_csv("DS1_C5_S5_Continuous_Practice_Computers_Data.csv")
df
```

```python
#a
df=df[df["price"]<=4000]
df
```

```python

u=df["screen"].unique()
u
```

```python
s=pd.DataFrame()
obj=df.groupby("screen")
for i in u:
    data=obj.get_group(i)
    data1=data.sample(20,random_state=78)
    s=s.append(data1,ignore_index=True)
s
```

```python
s.shape
```

```python
#b
```

```python
#population centraltendecny
mean1=df["price"].mean()
median1=df["price"].median()
mode1=df["price"].mode()[0]
std1=df["price"].std()
mean1,median1,mode1,std1
```

```python
#sample central tendency
mean=s["price"].mean()
median=s["price"].median()
mode=s["price"].mode()[0]
sd=s["price"].std()
mean,median,mode,sd
```

```python
#c distributions
```

```python
plt.title("population")
sb.kdeplot(df["price"],shade=True,color="Red")
```

```python
plt.title("sample")
sb.kdeplot(s["price"],shade=True,color="yellow")
```

```python
#sample central tendency is close to its population and we have similar distributions
```

```python
#e
s["zscore"]=(s["price"]-mean)/sd
s[["price","zscore"]]
```

```python
s["probability values of z scores"]=stat.norm.cdf(s["zscore"])
s[["price","zscore","probability values of z scores"]]
```
