```python
#Stratified sampling
#here we are grouping the data
#sampling method that involves the division of population into smaller subgroups known as Strata.
#it is done by dividing population into homogeneous group called Strata
#divides population into groups based on shared characteristics
#it captures key population characteristics .
#disadvantageous when we cant classify every member
```

```python
obj=df.groupby("Department")
obj
```

```python
u=df["Department"].unique()
u
```

```python
df3=pd.DataFrame()
for d in u:
    d_data=obj.get_group(d)#get_group constructs dataframe from group name provided
    sample_data=d_data.sample(5,random_state=34)#fetching 5 random employees department wise
    df3=df3.append(sample_data,ignore_index=True)
df3
```

```python
#Cluster sampling
```

```python
#In Cluster sampling, we divide the entire population into subgroups, wherein, each of those subgroups has similar characteristics to that of the population when considered in totality. Also, instead of sampling individuals, we randomly select the entire subgroups
```

```python
#method of obtaining represantative sample from a population
#an individual cluster is a subgroup that shares the same characteristics with other cluster
#use this approach when studying large,geographically dispersed population
#hard and also cost controlling process
#study characteristics of entire pop is impossible
#cluster sampling might be inly feasible method for obtaining a sample
#it increases the practicality of sampling from a large population
```

```python
#The company has offices in 10 cities across the country (all with roughly the same number of employees in similar roles). You donâ€™t have the capacity to travel to every office to collect your data, so you use random sampling to select 3 offices â€“ these are your clusters.
```

```python
u=["HR","IT"]
df4=pd.DataFrame()
for d in u:
    d_data=obj.get_group(d)#get_group constructs dataframe from group name provided
    sample_data=d_data.sample(frac=.2,random_state=34)
    #frac=.2 means 20% of each department data
    df4=df4.append(sample_data,ignore_index=True)
df4
```

```python
#Central limit theorem CLT
```

```python
#The CLT is a statistical theory that states that -
#The central limit theorem (CLT) states that the distribution of sample means approximates a normal distribution as the sample size gets larger.
#if you take a sufficiently large sample size from a population with a finite level of variance, 
#the mean of all samples from that population will be roughly equal to the population mean
```
