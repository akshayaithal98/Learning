```python
obj1=df.groupby("Year")
y=df["Year"].unique()
s4=pd.DataFrame()
y
```

# Task3.2

```python
s_10=df.sample(10,random_state=1)
s_20=df.sample(20,random_state=1)
s_30=df.sample(30,random_state=1)
s_40=df.sample(40,random_state=1)
```

```python
tb=pd.DataFrame()
```

```python
tb["s_10"]=s_10["Global_Sales_M$"].mean(),s_10["Global_Sales_M$"].median(),s_10["Global_Sales_M$"].std()
```

```python
tb["s_20"]=s_20["Global_Sales_M$"].mean(),s_20["Global_Sales_M$"].median(),s_20["Global_Sales_M$"].std()
```

```python
tb["s_30"]=s_30["Global_Sales_M$"].mean(),s_30["Global_Sales_M$"].median(),s_30["Global_Sales_M$"].std()
```

```python
tb["s_40"]=s_40["Global_Sales_M$"].mean(),s_40["Global_Sales_M$"].median(),s_40["Global_Sales_M$"].std()
```

```python
tb["population"]=df["Global_Sales_M$"].mean(),df["Global_Sales_M$"].median(),df["Global_Sales_M$"].std()
```

```python
tb["index"]=["mean","median","SD"]
```

```python
tb.set_index('index')
```

```python
plt.figure(figsize=(15,15))

plt.subplot(321)

plt.title("sample size 10")
sb.histplot(s_10["Global_Sales_M$"],bins=5,stat='probability',kde=True)

plt.subplot(322)
plt.title("sample size 20")
sb.histplot(s_20["Global_Sales_M$"],bins=5,stat='probability',kde=True)

plt.subplot(323)
plt.title("sample size 30")
sb.histplot(s_30["Global_Sales_M$"],bins=5,stat='probability',kde=True)

plt.subplot(324)
plt.title("sample size 40")
sb.histplot(s_40["Global_Sales_M$"],bins=5,stat='probability',kde=True)

plt.subplot(325)
plt.title("population")
sb.histplot(df["Global_Sales_M$"],bins=5,stat='probability',kde=True)

plt.show()
```

```python
#CONCLUSIONS
#As sample size increased from 10 to 40, sample mean became more closer to population mean.
#for sample size of 40 sample mean is nearly equivalent to population mean
#if we still increase sample size we may get even closer means and more normal distributed sample data
#but still from this observation we can say that mean gets closer to population mean as sample size gets larger which is CLT theorem.
```
